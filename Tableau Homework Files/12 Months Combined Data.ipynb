```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
load_data_201811 = "data/201811-citibike-tripdata.csv"
load_data_201812 = "data/201812-citibike-tripdata.csv"
load_data_201901 = "data/201901-citibike-tripdata.csv"
load_data_201902 = "data/201902-citibike-tripdata.csv"
load_data_201903 = "data/201903-citibike-tripdata.csv"
load_data_201904 = "data/201904-citibike-tripdata.csv"
load_data_201905 = "data/201905-citibike-tripdata.csv"
load_data_201906 = "data/201906-citibike-tripdata.csv"
load_data_201907 = "data/201907-citibike-tripdata.csv"
load_data_201908 = "data/201908-citibike-tripdata.csv"
load_data_201909 = "data/201909-citibike-tripdata.csv"
load_data_201910 = "data/201910-citibike-tripdata.csv"

# Read the City and Ride Data
df_201811 = pd.read_csv(load_data_201811)
df_201812 = pd.read_csv(load_data_201812)
df_201901 = pd.read_csv(load_data_201901)
df_201902 = pd.read_csv(load_data_201902)
df_201903 = pd.read_csv(load_data_201903)
df_201904 = pd.read_csv(load_data_201904)
df_201905 = pd.read_csv(load_data_201905)
df_201906 = pd.read_csv(load_data_201906)
df_201907 = pd.read_csv(load_data_201907)
df_201908 = pd.read_csv(load_data_201908)
df_201909 = pd.read_csv(load_data_201909)
df_201910 = pd.read_csv(load_data_201910)
```

```python
result = pd.concat([df_201811,df_201812,df_201901,df_201902,df_201903,df_201904,df_201905,
                    df_201906,df_201907,df_201908,df_201909,df_201910], axis=0, join='outer', ignore_index=False)
```

```python
result.head()
```

```python
result.loc[result['gender'] == 2,["gender"]] = 'Female'
```

```python
result.loc[result['gender']==1, ['gender']] = 'male'
```

```python
result.loc[result['gender'] == 0,["gender"]] = 'Unknown'
```
